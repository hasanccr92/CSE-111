```python
'''1. Create a class called Calculator.
2. Your class shall have 1 constructor and 4 methods, namely add, subtract, multiply and divide.
3. Now, create an object of your class. After creating an object, it should print “Let’s Calculate!”
4. Then take 3 inputs from the user: first value, operator, second value
5. Now based on the given operator, call the required method and print the result.'''
class Calculator:
    def __init__(self,n1,op,n2):
        self.n1 = n1
        self.op = op
        self.n2 = n2
        print("Let's Calculate!")
        print("Value 1:",self.n1)
        print("Operator:",self.op)
        print("Value 2:",self.n2)
    def add(self):
            print("Result:",self.n1 + self.n2)
    def sub(self):
            print("Result:",self.n1 - self.n2)
    def mul(self):
            print("Result:",self.n1 * self.n2)
    def div(self):
            print("Result:",self.n1 / self.n2)
n1 = int(input("Enter value 1: "))
op = input("Enter operator: ")
n2 = int(input("Enter value 2: "))
#output = Calculator(n1,op,n2)
if op == "+":
    output = Calculator(n1,op,n2).add()
if op == "-":
    output = Calculator(n1,op,n2).sub()
if op == "*":
    output = Calculator(n1,op,n2).mul()
if op == "/":
    output = Calculator(n1,op,n2).div()
```

```python
#2
class Customer:
    def __init__(self,name):
        self.name = name
    def greet(self,name1=""):
        print("Hello!",name1)
    def purchase(self,*items):
        self.items = items
        itm = 0
        for i in items:
            itm += 1
        print(self.name,",you have purchased",itm,"item(s)")
        for i in self.items:
            print(i)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#8
class Student:
    def __init__(self,name,id,dept="CSE"):
        self.name = name
        self.id = id
        self.dept = dept
    def dailyEffort(self,hr):
        self.hr = hr
        if hr<=2:
            self.sug = ('Suggestion: Should give more effort!')
        elif hr <= 4:
            self.sug = ("Suggestion: Keep up the good work!")
        else:
            self.sug = ('Suggestion: Excellent! Now motivate others.')
    def printDetails(self):
        print("Name:",self.name)
        print("ID:",self.id)
        print("Department:",self.dept)
        print("Daily Effort:",self.hr,"hour(s)")
        print(self.sug)

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#11
class Shinobi:
    def __init__(self,name,rank):
        self.name = name
        self.rank = rank
        self.mission = 0
        self.salary = 0
    def changeRank(self,rank):
        self.rank = rank
    def calSalary(self,mission):
        self.mission = mission
        if self.rank == "Genin":
            self.salary = mission*50
        elif self.rank == "Chunin":
            self.salary = mission * 100
        else:
            self.salary = mission*500
    def printInfo(self):
        print("Name:",self.name)
        print("Rank:",self.rank)
        print("Number of mission:",self.mission)
        print("Salary:",self.salary)

naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
#12
class ParcelKoro:
    def __init__(self,name="No name set",weight=0,fee=0):
        self.name = name
        self.product_weight = weight
        self.fee = fee    
    def calculateFee(self,loc=""):
        self.loc = loc
        if self.product_weight > 0:
            self.fee += (self.product_weight * 20)
            if loc == "":
                self.fee +=50
            else:
                self.fee +=100

        else:
            self.fee += 0

    def printDetails(self):
        print("Customer Name:",self.name)
        print("Product weight:",self.product_weight)
        print("Total fee:",self.fee)

print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#13
class Batsman:
    def __init__(self,*details):
        if len(details) == 3:
            self.name = details[0]
            self.run = details[1]
            self.ball = details[2]
        elif len(details) == 2:
            self.name = "New Batsman"
            self.run = details[0]
            self.ball = details[1]
    def printCareerStatistics(self):
        print("Name:",self.name)
        print("Runs scored:",self.run," Balls faced:",self.ball)
        return ""

    def battingStrikeRate(self):
        sr = self.run/(self.ball)*100
        print(sr)
        return ""


    def setName(self,name):
        self.name = name


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#14
class EPL_Team:
    def __init__(self,name,song="No Slogan",title=0):
        self.name = name
        self.song = song
        self.title = title
    def increaseTitle(self):
        self.title +=1
    def changeSong(self,song):
        self.song = song

    def showClubInfo(self):
        return f"Name:{self.name} \nSong:{self.song} \nTotal No of title:{self.title}"

manu = EPL_Team('Manchester United', 'Glory Glory Man United')
chelsea = EPL_Team('Chelsea')
print('===================')
print(manu.showClubInfo())
print('##################')
manu.increaseTitle()
print(manu.showClubInfo())
print('===================')
print(chelsea.showClubInfo())
chelsea.changeSong('Keep the blue flag flying high')
print(chelsea.showClubInfo())
```

```python
#15
class Account:
    def __init__(self,name="Default Account",balance=0.0):
        self.name = name
        self.balance = balance
    def details(self):
        return f"{self.name}\n{self.balance}"
    def withdraw(self,money):
        self.money = money
        if (self.balance-self.money) <= 3070:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum")
        else:
            print("Withdraw successful! New balance is:",self.balance-self.money)
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)
```
