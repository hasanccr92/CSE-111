```python
#1
class Student:
    id = 1
    def __init__(self,name,dept,age,cgpa):
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cgpa
        self.id = Student.id
        Student.id += 1

    def get_details(self):
        print('ID:',self.id,"\nName:",self.name,"\nDepartment:",self.dept,"\nAge:",self.age,"\nCGPA:",self.cgpa)
    @classmethod
    def from_String(cls,info):
        name,dept,age,cgpa = info.split("-")
        return cls(name,dept,age,cgpa)

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print('Subtask5:\nClass variables are shared in all objects and instance variables are different for each object. Classs variables are called outside the design block and they are same for all objects. Instance variables are designated inside the init method and not same for all objects.')
print('Subtask6:\nAn instance method belongs to objects of a class, not to the class and a class method belongs to a class and its parameter is always the class itself.')
```

```python
#2
class Assassin:
    anum = 1
    def __init__(self,name,sucrt=0):
        self.name = name
        self.sucrt = sucrt
        self.anum = Assassin.anum
        Assassin.anum += 1
    def printDetails(self):
        print("Name:",self.name,"\nSuccess rate:",self.sucrt,"%\nTotal number of Assassin:",self.anum)

    @classmethod
    def failureRate(cls,name,frate):
        odj = cls(name,100-frate)
        return odj
    @classmethod
    def failurePercentage(cls,name,fprtg):
        obj = cls(name,100-fprtg)
        return obj

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#3
class Passenger:
    count = 0
    def __init__(self,name):
        self.name = name
        self.busfare = 450
        Passenger.count += 1

    def set_bag_weight(self,weight):
        self.weight = weight
    def printDetail(self):
        if 50 > self.weight > 20:
            self.busfare += 50
        if self.weight > 50:
            self.busfare += 100
        if 20 > self.weight > 0:
            self.busfare += 0
        print("Name:",self.name,"\nBus Fare:",self.busfare,'taka')

print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#5
class Employee:
    def __init__(self,name,workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
    def name(self):
        return self.name
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        obj = cls(name,2021-year)
        return obj
    @staticmethod
    def experienceCheck(year,gender):
        if gender == 'male' and year<3:
            return "He is not experienced"
        elif gender == 'male' and year>3:
            return "He is experienced"
        if gender == 'female' and year<3:
            return "She is not experienced"
        else: 
            return "She is experienced"

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#6
class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount = Laptop.laptopCount + count
    @staticmethod
    def advantage():
        print("Laptops are portable")
    @classmethod
    def resetCount(cls):
        cls.laptopCount = 0

lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#7
class Cat:
    Number_of_cats = 0
    def __init__(self,color,action):
        self.color = color
        self.action = action
        Cat.Number_of_cats += 1
    def printCat(self):
        print(f'{self.color} cat is {self.color}')
    @classmethod
    def no_parameter(cls,color = 'White',action = 'sitting'):
        obj = cls(color,action)
        return obj
    @classmethod
    def first_parameter(cls,color,action = 'sitting'):
        obj = cls(color,action)
        return obj
    @classmethod
    def second_parameter(cls,action,color = 'Grey'):
        obj = cls(color,action)
        return obj
    def changeColor(self,clr):
        self.color = clr


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#8
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        print("Default radius=", Cylinder.radius, "and height=", Cylinder.height)
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print("Updated radius=",Cylinder.radius,"and height=",Cylinder.height)
    @staticmethod
    def area(radius, height):
        area = 2*math.pi*radius*height + 2*math.pi*(radius ** 2)
        print("Area:",area)
    @staticmethod
    def volume(radius, height):
        volume = (radius ** 2)*height*math.pi
        print("Volume:",volume)
    @classmethod
    def swap(cls,radius1,height1):
        obj1 = cls(height1,radius1,)
        return obj1
    @classmethod
    def changeFormat(cls,info):
        radius2,height2 = info.split("-")
        obj2 = cls(float(radius2),float(height2))
        return obj2

    
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#10
class SultansDine:
    total_branch = 0
    total_sell = 0
    list = []
    def __init__(self, loc):
        self.loc = loc
        SultansDine.total_branch += 1
        SultansDine.list.append(loc)
    def sellQuantity(self, qty):
        if qty < 10:
            self.qty = qty * 300
        elif qty < 20:
            self.qty = qty * 350
        else:
            self.qty = qty * 400

        self.Branch_sell = self.qty
        SultansDine.list.append(self.Branch_sell)
        SultansDine.total_sell += self.Branch_sell
        self.sell_percentage = (self.Branch_sell / SultansDine.total_sell) * 100

    def branchInformation(self):
        print(f'Branch Name: {self.loc}')
        print(f'Branch Sell: {self.Branch_sell} Taka')
    @classmethod
    def details(cls):
        print(f'Total Number of branch(s): {SultansDine.total_branch}')
        print(f'Total Sell: {SultansDine.total_sell} Taka')
        length = len(SultansDine.list)
        for i in range(0, length, 2):
            sell_percentage = (SultansDine.list[i + 1] / SultansDine.total_sell) * 100
            new_sell_percentage = str(round(sell_percentage, 2))
            
            print(f'Branch Name: {SultansDine.list[i]}, Branch Sell: {SultansDine.list[i + 1]} Taka')
            print(f"Branch consists of total sell's: {new_sell_percentage}%")


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
